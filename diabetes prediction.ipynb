```python
from sklearn.datasets import load_diabetes
```

```python
datasets=load_diabetes()
```

```python
datasets
```

```python
print(datasets['DESCR'])
```

```python
import pandas as pd 
df_diabetes=pd.DataFrame(datasets.data,columns=[
    'age',
    'sex',
    'bmi', 
    'bp', 
    's1',
    's2',
    's3', 
    's4', 
    's5', 
    's6'
    
    
])
df_diabetes.head()
```

```python
x=df_diabetes
y=datasets['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.head()
```

```python
x_train.corr()
```

```python
from sklearn.tree import  DecisionTreeRegressor
regressor=DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

```python
## hyper paramter tunning
```

```python
param= {
    "criterion":['squared_error','friedman_mse',"absolute_error"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5,10,15,20,25], 
    "max_features":['auto','sqrt','log2']
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring='neg_mean_squared_error')
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
yp=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error,mean_squared_error
```

```python
print(r2_score(y_test,yp))
print(mean_absolute_error(y_test,yp))
print(mean_squared_error(y_test,yp))
```

```python
selectmodel=DecisionTreeRegressor(criterion='friedman_mse',max_depth=4,max_features='log2',splitter='random')
```

```python
selectmodel.fit(x_train,y_train)
```

```python
import matplotlib.pyplot as plt 
from sklearn import tree
plt.figure(figsize=(15,10))
tree.plot_tree(selectmodel,filled=True)
```

## datsets cleaning

```python
dataset[dataset.isnull().any(axis=1)]
```

```python


dataset.loc[:121, "Region"] = 0
dataset.loc[122:, "Region"] = 1
df = dataset
```

```python
df.info()
```

```python
df.head()
```

```python
df["Region"]=df[["Region"]].astype(int)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.info()
```

## change the required columns as integer data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

## changing other colummns to float datatype

```python
objects= [features for features in df.columns if df[features].dtypes=='O']
```

```python
objects
```

```python
for i in objects:
    if i !='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

```python
df.to_csv("Algerian_cleaned_datasets.csv",index=False)
```

## Exploratory data analysis

```python
# drop day month and year

df_copy=df.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
plt.style.use('seaborn-v0_8')
df_copy.hist(bins=50,figsize=(20,15))
plt.show()
```

```python
percentage=df_copy['Classes'].value_counts(normalize=True)*100
```

```python
claslabels=['fire','Not fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=claslabels,autopct='%1.1f%%')
plt.show()
```

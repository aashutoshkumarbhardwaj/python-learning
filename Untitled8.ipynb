## Support vector implementation

```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=2,n_classes=2,n_clusters_per_class=2,n_redundant=0)
```

```python
x
```

```python
sns.scatterplot(x=pd.DataFrame(x)[0],y=pd.DataFrame(x)[1],hue=y)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.svm import SVC
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train,y_train)
```

```python
yp=svc.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,yp))
print(confusion_matrix(y_test,yp))
```

```python
rbf=SVC(kernel='rbf')
rbf.fit(x_train,y_train)
```

```python
yp1=rbf.predict(x_test)
```

```python
print(classification_report(y_test,yp1))
print(confusion_matrix(y_test,yp1))
```

```python
polynomial=SVC(kernel='poly')
polynomial.fit(x_train,y_train)

```

```python
yp2=polynomial.predict(x_test)
```

```python
print(classification_report(y_test,yp2))
print(confusion_matrix(y_test,yp2))
```
